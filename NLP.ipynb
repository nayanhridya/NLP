### **Q1. Text Processing (Basics)**
* a. Define a string containing a paragraph as the value.
* b. Write a program to print the number of total words and total unique words in the
paragraph.
* c. Find the frequency of all words and also display the most and least frequent word.
* d. Find the longest word in the paragraph.

```python
paragraph = """
Crop disease prediction and AI-driven agriculture systems are revolutionizing modern farming.
By leveraging machine learning and data analytics, these systems can predict the onset of diseases,
optimize crop yields, and reduce the use of pesticides. Remote sensing technologies, such as drones and satellites,
provide real-time data on crop health. This information, combined with weather data and soil conditions,
enables farmers to make informed decisions. AI algorithms analyze vast amounts of data to identify patterns and
predict potential outbreaks, ensuring timely interventions. Such systems enhance productivity, sustainability, and profitability in agriculture.
"""
```

```python
import re
from collections import Counter

# Clean and split the paragraph into words
words = re.findall(r'\b\w+\b', paragraph.lower())

# Total number of words
total_words = len(words)

# Total number of unique words
unique_words = len(set(words))

print(f"Total words: {total_words}")
print(f"Total unique words: {unique_words}")
```

```python
# Calculate word frequencies
word_freq = Counter(words)

# Most frequent word
most_frequent_word = word_freq.most_common(1)[0]

# Least frequent word
least_frequent_word = word_freq.most_common()[-1]

print(f"Word frequencies: {word_freq}")
print(f"Most frequent word: {most_frequent_word}")
print(f"Least frequent word: {least_frequent_word}")
```

```python
# Longest word in the paragraph
longest_word = max(words, key=len)

print(f"Longest word: {longest_word}")
```

### **Q2. Regular Expression**
* Solve exercise 2.1 and 2.2 from Text book of Speech and Language Processing of Daniel
Jurafsky and team

```python
# Alphabetic strings example
alphabetic_strings = "Crop disease prediction and AI-driven agriculture systems"

# Regular expression for alphabetic strings
regex_alpha = r'[a-zA-Z]+'

# Find all matches
matches_alpha = re.findall(regex_alpha, alphabetic_strings)
print(f"Alphabetic strings: {matches_alpha}")

# Lowercase strings ending in 'b' example
lowercase_b = "herb"

# Regular expression for lowercase strings ending in 'b'
regex_lower_b = r'^[a-z]*b$'

# Find all matches
matches_lower_b = re.findall(regex_lower_b, lowercase_b, re.MULTILINE)
print(f"Lowercase strings ending in 'b': {matches_lower_b}")

# Strings where each 'a' is surrounded by 'b's example
ab_strings = "bab abba bababa bbabab"

# Regular expression for 'a' surrounded by 'b's
regex_ab = r'^(b*(ab)+b*)$'

# Find all matches
matches_ab = re.findall(regex_ab, ab_strings, re.MULTILINE)
print(f"Strings with 'a' surrounded by 'b': {matches_ab}")

# Exercise 2.2: Write regular expressions for specified languages

# Strings with two consecutive repeated words example
repeated_words = "Humbert Humbert the the crop crop disease"

# Regular expression for consecutive repeated words
regex_repeated = r'\b(\w+)\s+\1\b'

# Find all matches
matches_repeated = re.findall(regex_repeated, repeated_words)
print(f"Consecutive repeated words: {matches_repeated}")

# Example string
start_int_end_word = "2024 wheat"

# Regular expression for strings starting with an integer and ending with a word
regex_int_word = r'^\d+.*\b[a-zA-Z]+\b$'

# Match the string
match_int_word = re.match(regex_int_word, start_int_end_word)
if match_int_word:
    print(f"String that starts with an integer and ends with a word: {start_int_end_word}")
else:
    print("No match found.")

# Example string
grotto_raven = "The mysterious grotto was the home of the raven"

# Regular expression for strings with both 'grotto' and 'raven'
regex_grotto_raven = r'\b(?:grotto\b.*\braven|raven\b.*\bgrotto)\b'

# Match the string
match_grotto_raven = re.search(regex_grotto_raven, grotto_raven)
if match_grotto_raven:
    print(f"String with both 'grotto' and 'raven': {grotto_raven}")
else:
    print("No match found.")

# Example string
sentence = "Crop disease prediction is crucial."

# Regular expression to capture the first word
regex_first_word = r'^\b(\w+)\b(.*)$'

# Match the string
match_first_word = re.match(regex_first_word, sentence)
if match_first_word:
    first_word = match_first_word.group(1)
    rest_sentence = match_first_word.group(2)
    print(f"First word: {first_word}")
    print(f"Rest of the sentence: {rest_sentence}")
else:
    print("No match found.")
```

```python
import re
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Define the chapter content
chapter_content="""In the fertile fields of Agriville, farmers have long relied on their instincts and traditional methods to ensure bountiful harvests. However, as the climate began to change and new crop diseases emerged, they faced unprecedented challenges. This is where Dr. Elena Shaw, an innovative agronomist, came into the picture with her groundbreaking AI-driven agriculture system.

Dr. Shaw had spent years developing an advanced AI model capable of predicting crop diseases before they could wreak havoc on the fields. She named her creation AGRI-Guardian. AGRI-Guardian was equipped with sensors placed throughout the fields, gathering data on soil moisture, temperature, humidity, and plant health. This data was then analyzed in real-time by the AI, which used sophisticated algorithms to identify patterns and predict potential disease outbreaks.

One sunny morning, Dr. Shaw received an alert from AGRI-Guardian. The system had detected unusual patterns in the data collected from a section of wheat fields. The AI predicted a potential outbreak of rust disease, a devastating fungal infection that could ruin the crop. Dr. Shaw immediately informed the farmers, who took preventive measures by applying the recommended fungicides.

Thanks to AGRI-Guardian’s timely warning, the farmers managed to save their crops from the impending threat. The news of this success spread quickly, and farmers from neighboring regions became eager to adopt the AI-driven system.

Dr. Shaw continued to refine AGRI-Guardian, incorporating feedback from the farmers and advancing its predictive capabilities. She also added features to recommend the best times for planting and harvesting, tailored to each farm’s unique conditions. The system even provided insights into optimal crop rotation practices, helping to maintain soil fertility and reduce the risk of disease.

In a particularly dry season, AGRI-Guardian played a crucial role in water management. By analyzing weather forecasts and soil moisture levels, it advised farmers on precise irrigation schedules, ensuring that every drop of water was used efficiently. This not only conserved water but also prevented over-irrigation, which could lead to root diseases.

One day, a young farmer named Raj, who had recently adopted AGRI-Guardian, noticed a significant improvement in his crop yields. He was able to produce healthier crops with fewer resources, leading to higher profits. Raj became an advocate for AI-driven agriculture, encouraging other young farmers to embrace technology and modernize their practices.

As the seasons passed, the community of Agriville thrived. The fields flourished with healthy crops, and the farmers gained newfound confidence in their ability to tackle challenges. AGRI-Guardian had transformed their way of life, bringing them into a new era of sustainable and efficient farming.

Dr. Shaw’s vision of merging technology with traditional farming practices had become a reality. Her AI-driven agriculture system not only protected crops from diseases but also empowered farmers with knowledge and tools to make informed decisions. The story of AGRI-Guardian spread far and wide, inspiring agronomists and farmers alike to explore the endless possibilities of AI in agriculture.

And so, Agriville continued to prosper, a testament to the power of innovation and the resilience of the farming community. With AGRI-Guardian by their side, the farmers of Agriville were ready to face whatever challenges the future might bring, confident in their ability to grow and sustain their land for generations to come.
"""

# Function to clean and split the text into words
def get_words(text):
    return re.findall(r'\b\w+\b', text.lower())

# Function to split the text into sentences
def get_sentences(text):
    return re.split(r'(?<=[.!?]) +', text.strip())

# Function to get word frequencies
def get_word_frequencies(words):
    return Counter(words)

# Get words and sentences
words = get_words(chapter_content)
sentences = get_sentences(chapter_content)

# Total number of words
total_words = len(words)

# Total number of unique words
unique_words = len(set(words))

# Total number of sentences
total_sentences = len(sentences)

# Average words per sentence
average_words_per_sentence = total_words / total_sentences

# Word frequencies
word_frequencies = get_word_frequencies(words)

# Most and least frequent words
most_frequent_word = word_frequencies.most_common(1)[0]
least_frequent_word = word_frequencies.most_common()[-1]

# Longest word
longest_word = max(words, key=len)

# Display results
print(f"Total words: {total_words}")
print(f"Total unique words: {unique_words}")
print(f"Total sentences: {total_sentences}")
print(f"Average words per sentence: {average_words_per_sentence:.2f}")
print(f"Most frequent word: {most_frequent_word}")
print(f"Least frequent word: {least_frequent_word}")
print(f"Longest word: {longest_word}")

# Plot word frequencies
top_n = 10
most_common_words = word_frequencies.most_common(top_n)
words, frequencies = zip(*most_common_words)

# Enhancing the bar plot
plt.figure(figsize=(12, 8))
plt.bar(words, frequencies, color='skyblue')
plt.xlabel('Words', fontsize=14)
plt.ylabel('Frequencies', fontsize=14)
plt.title('Top 10 Most Common Words', fontsize=16)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Word cloud for visualizing word frequencies
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_frequencies)

plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of the Chapter', fontsize=16)
plt.show()

# Sentence length distribution
sentence_lengths = [len(get_words(sentence)) for sentence in sentences]

plt.figure(figsize=(12, 8))
plt.hist(sentence_lengths, bins=range(0, max(sentence_lengths) + 1, 2), color='skyblue', edgecolor='black')
plt.xlabel('Sentence Length (in words)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.title('Sentence Length Distribution', fontsize=16)
plt.xticks(range(0, max(sentence_lengths) + 1, 2), fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
